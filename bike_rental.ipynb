```python
data.drop(['instant','hr'],axis=1,inplace=True)
```

```python
data
```

```python
data['dteday']=pd.to_datetime(data['dteday'])
```

```python
data
```

```python
daily_data = data.groupby('dteday').agg({
    'season': 'first',        
    'yr': 'first',            
    'mnth': 'first',          
    'holiday': 'first',       
    'weekday': 'first',       
    'workingday': 'first',    
    'weathersit': 'mean',     
    'temp': 'mean',          
    'atemp': 'mean',          
    'hum': 'mean',            
    'windspeed': 'mean',      
    'casual': 'sum',          
    'registered': 'sum',      
    'cnt': 'sum'              
}).reset_index()
```

```python
daily_data
```

```python
daily_data.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

plt.figure(figsize=(8, 6))
sns.boxplot(y='cnt', data=daily_data)
plt.title('Outliers in Total Bike Rentals')
plt.show()
```
